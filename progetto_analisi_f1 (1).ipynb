APERTURA E LETTURA FILE CSV

```python
import csv

dati = []

with open('/content/drive/MyDrive/progetto Python /formula1_data.csv') as csv_file:
  csv_reader = csv.DictReader(csv_file)

  for riga in csv_reader:
    dati.append(riga)

# dopo aver montato il drive, importo il file csv come dizionario e lo trasformo in una lista di dizionari, cosi da poterlo leggere in tutte le funzioni.
```

FUNZIONE N. 1 ANALISI PERFORMANCE PILOTI

```python
def performance(pilota):

  punti = 0
  vittorie = 0
  podi = 0

  for riga in dati:
    if riga['Driver'] == pilota:
      posizione = int(riga['Position'])

      if posizione == 1:
        punti += 10
        vittorie += 1
        podi += 1
      elif posizione == 2:
        punti += 8
        podi += 1
      elif posizione == 3:
        punti += 6
        podi += 1
      elif posizione == 4:
        punti += 5
      elif posizione == 5:
        punti += 4
      elif posizione == 6:
        punti += 3
      elif posizione == 7:
        punti += 2
      elif posizione == 8:
        punti += 1

  return punti, vittorie, podi

pilota = input('Pilota: ')
punti, vittorie, podi = performance(pilota)

print(f'Punti: {punti}')
print(f'Vittorie: {vittorie}')
print(f'Podi: {podi}')

# creo la funzione partendo dalle tre variabili richieste. con il ciclo for (se il nome del pilota corrisponde) controllo ogni riga della lista, calcolando i punteggi in base alla posizione
# definisco la funzione e infine  stampo i risultati
```

FUNZIONE N. 2 CLASSIFICA PILOTI IN FILE .TXT

```python
def classifica():

  punteggi = {}

  for riga in dati:
    pilota = riga['Driver']
    posizione = int(riga['Position'])

    if posizione == 1:
      punti = 10
    elif posizione == 2:
      punti = 8
    elif posizione == 3:
      punti = 6
    elif posizione == 4:
      punti = 5
    elif posizione == 5:
      punti = 4
    elif posizione == 6:
      punti = 3
    elif posizione == 7:
      punti = 2
    elif posizione == 8:
      punti = 1
    else:
      punti = 0

    if pilota in punteggi:
      punteggi[pilota] += punti
    else:
      punteggi[pilota] = punti

  return punteggi

with open('Drivers_Standings_2008.txt', 'w') as file:
  file.write('Drivers Standings 2008:\n\n')
  punteggi = classifica()
  for pilota in punteggi:
    punti = punteggi[pilota]
    file.write(f'{pilota}: {punti}\n')

# in questa funzione creo un dizionario (inizialmente vuoto) dove per ogni riga della lista vado a calcolare il punteggio in base alla posizione.
# poi utilizzo if: se il pilota è presente nel dizionario si sommano i punti, altrimenti si crea una nuova chiave con i valori assegnati.
# infine scrivo la classifica nel file txt stampando nome e punti.
```

FUNZIONE N. 3 CLASSIFICA COSTRUTTORI

```python
def class_costruttori():
  costruttori = {}

  for riga in dati:
    team = riga['Team']
    posizione = int(riga['Position'])

    if posizione == 1:
      punti = 10
    elif posizione == 2:
      punti = 8
    elif posizione == 3:
      punti = 6
    elif posizione == 4:
      punti = 5
    elif posizione == 5:
      punti = 4
    elif posizione == 6:
      punti = 3
    elif posizione == 7:
      punti = 2
    elif posizione == 8:
      punti = 1
    else:
      punti = 0

    if team in costruttori:
      costruttori[team] += punti
    else:
      costruttori[team] = punti

  return costruttori

costruttori = class_costruttori()

class_ordinata = dict(sorted(costruttori.items(), key=lambda item: item[1], reverse=True))

for team, punti in class_ordinata.items():
  print(f'team {team}: {punti}')

# in questa funzione creo un dizionario (inizialmente vuoto) dove per ogni riga della lista vado a calcolare i punteggi.
# anche in questo caso se il team è presente sommo il punteggio, altrimenti crea una nuova chiave con i valori assegnati.
# -- la variabile class_ordinata mi è stata proposta in automatico da Colab, e l'ho presa per buona perchè non riuscivo ad ordinare la classifica --
# infine per ogni riga della classifica stampo nome e i punti.
```
